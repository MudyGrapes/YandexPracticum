### 2. Изменение продаж по платформам. Когда появляются новые и исчезают старые?

```python
platform_sales = (df
                  .pivot_table(index='platform', 
                                 values='total_sales', 
                                 aggfunc='sum')
                  .sort_values(by='total_sales', ascending=False)
                  .reset_index())
print(platform_sales)

#Ограничимся PS4, т.к. Game Boy - это уже почти ретро даже для 2016го
top_platforms = list(platform_sales['platform'][:7])
top_platforms.remove('PS') #Конечно же уберем PS1, ведь на дворе 2016
#Построим гистограмму продаж игр по платформам по годам
def plot_platform_hist(top_platforms):
    """
    Построить гистограмму выпуска игр на платформу по годам
    top_platforms -- список платформ (list)
    """
    for platform in top_platforms:
        print(platform)
        platform_years = df[df['platform']==platform]['year_of_release']
        max_year, min_year = max(platform_years), min(platform_years)
        platform_years.plot(kind='hist', 
                            alpha=0.4,
                            label=platform,
                            bins=max_year-min_year)
    plt.xlim(2000, 2017)
    plt.legend(loc="upper left") 
    plt.show()
plot_platform_hist(top_platforms[:3])
plot_platform_hist(top_platforms[3:6])
plot_platform_hist(['PS2', 'Wii', 'DS'])
#Отдельно глянем PC в сравнении
plot_platform_hist(['PS4', 'PC', 'X360'])
```

<div style="border:solid blue 1px; padding: 20px">
    
* Как видно в топах по продажам 7 платформ
* Но брать PS1 для 2016 года совсем нерелевантно
* К тому же видим как к 2011 падает до нуля PS2, к 2013 NintendoDS (оно и логично когда уже давным давно смартфоны есть), а под Wii почти ничего не выпускают с 2013
* Очевидно, что "флагманы" в 2016 году это PS3, PS4 и X360
* Также не стоит упускать из виду возросшую популярность PC игр

### 5. Какие платформы лидируют по продажам, растут или падают?

```python
platform_sales_by_years = (df_actual
                          .pivot_table(index='year_of_release', 
                                       columns='platform',
                                       values='total_sales', 
                                       aggfunc='sum')
                          .reset_index())

print(platform_sales_by_years)

def plot_platform_sales_hist(plot_platforms):
    """
    Построить график общих продаж игр на платформу по годам
    plot_platforms -- список платформ (list)
    """
    (platform_sales_by_years[['year_of_release'] + plot_platforms]
     .plot
     .area(x='year_of_release', stacked=False, figsize=(10, 4)))
    plt.title('total_sales')
    plt.xlim(2013, 2017)
    plt.ylim(0, 120)
    plt.xticks(actual_years)
    plt.show()
plot_platform_sales_hist(['PS4', 'PS3', 'X360', 'PC'])
plot_platform_sales_hist(['3DS', 'DS', 'PSP', 'PSV'])
plot_platform_sales_hist(['Wii', 'WiiU', 'XOne'])
```

```python
#Популярная платформа - это PS4
def sales_score_correlation(platform_list):
    """
    Посчитать корреляцию и построить график связи продаж и оценок критиков и пользователей
    platform -- список платформ (list)
    """
    for platform in platform_list:
        df_actual_platform = df_actual[df_actual['platform']==platform]
        sales_critic_corr = df_actual_platform['total_sales'].corr(df_actual_platform['critic_score'])
        sales_user_corr = df_actual_platform['total_sales'].corr(df_actual_platform['user_score'])
        if not math.isnan(sales_critic_corr):
            plt.scatter(df_actual_platform['total_sales'], df_actual_platform['critic_score'])
            plt.title('Корреляция между продажами и оценками критиков '+platform+': ' + 
                      str(sales_critic_corr))
            plt.ylabel('total_sales, mln pcs')
            plt.xlabel('critic_score')
            plt.show()
        if not math.isnan(sales_user_corr):
            plt.scatter(df_actual_platform['total_sales'], df_actual_platform['user_score'])
            plt.title('Корреляция между продажами и оценками пользователей '+platform+': ' + 
                      str(sales_user_corr))
            plt.ylabel('total_sales, mln pcs')
            plt.xlabel('user_score')
            plt.show()
sales_score_correlation(['PS4'])
```

## Шаг 5. Проверка гипотез

<a id='18-bullet'></a>

### 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

```python
#Посчитаем среднее и дисперсию
user_scores_pc = df_actual.query('platform=="PC" and not user_score.isna()')['user_score']
user_scores_xone = df_actual.query('platform=="XOne" and not user_score.isna()')['user_score']
print(df_actual
      .query('platform in ["XOne", "PC"]')
      .pivot_table(index='platform', 
                   values='user_score',
                   aggfunc=[np.mean, np.var, np.std, 'count']), '\n')
#Проверим гипотезу о случайности неравенстве дисперсий тестом Levene
stat, p_levene = st.levene(user_scores_pc, user_scores_xone)
print('Случайность отличия дисперсий: p_levene)

#Но - средние одинаковые - различия случайны
#Н1 - средние неравны - различия неслучайны
# Т.к. дисперсии отличаются необходимо принять параметр equal_var=False (t-тест Уэлча)
# Уровень значимости примем 5%
alpha = 0.05
results = st.ttest_ind(user_scores_pc, user_scores_xone, equal_var=False)
print(results)
print('p-value:', results.pvalue, '\n')
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
    
user_scores_pc.plot(kind='kde',label='PC')
user_scores_xone.plot(kind='kde',title="user_scores_density",label='XOne',figsize=(11,6))
plt.legend(loc="upper right") 
```
